# Praca domowa 5

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/mini-pw/2021L-WUM/main/Prace_domowe/Praca_domowa5/clustering.csv",
                  header=None)
```

```python
data.columns = ["x", "y"]
data.head()
```

```python
plt.scatter(data["x"], data["y"])
plt.title("Data")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
# nie ma wyraźnych outlierów
```

## kmeans

```python
visualizer = KElbowVisualizer(KMeans(random_state=0), k=(1,21))
visualizer.fit(data)
visualizer.show()
# disortion - suma kwadratów odległości poszczególnych obserwacji od średniej w ich klastrze
```

```python
kmeans = KMeans(random_state=0, n_clusters=5)
kmeans.fit(data)
sns.scatterplot(x="x", y="y", data=data, hue=kmeans.labels_, legend="full")
plt.title("Data clustered by KMeans")
plt.show()
# w miarę sensowny podział,
# ale te kilka obserwacji nad i po prawej od klastra 0
# może lepiej byłoby przypisać do niego
```

## agglomerative

```python
agg_clustering = AgglomerativeClustering(n_clusters=6)
agg_clustering.fit(data)
```

```python
sns.scatterplot(x="x", y="y", data=data, hue=agg_clustering.labels_, legend="full")
plt.title("Data clustered by agglomerative")
plt.show()
```
